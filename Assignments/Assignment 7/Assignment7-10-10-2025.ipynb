```python
# Importing required libraries
import matplotlib.pyplot as plt

# Sample data
x = [10, 20, 30, 40, 50]
y = [5, 15, 25, 35, 45]

# Create a line plot
plt.plot(x, y, color='magenta', marker='s', linestyle='-.', label='Line Plot')

# Adding title and labels
plt.title("Modified Line Plot")
plt.xlabel("New X-axis")
plt.ylabel("New Y-axis")
plt.legend()  # Show legend
plt.grid(True)  # Add grid
plt.show()
```

```python
# Data
categories = ['English', 'History', 'Geography', 'Economics']
marks = [65, 70, 60, 75]

# Create bar plot
plt.bar(categories, marks, color='coral')

# Add title and labels
plt.title("Marks of Students in New Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
# Sample data
ages = [25, 26, 27, 25, 28, 29, 27, 26, 28, 29, 30, 28]

# Create histogram
plt.hist(ages, bins=4, color='teal', edgecolor='white')

# Add title and labels
plt.title("New Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Sample data
x = [3, 6, 9, 6, 1, 12, 1, 8]
y = [55, 66, 77, 88, 99, 66, 77, 88]

# Create scatter plot
plt.scatter(x, y, color='navy', marker='^')
plt.title("Modified Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

# Seaborn

```python
# Seaborn library
import seaborn as sns
import pandas as pd

# Sample dataset
data = {'Student': ['E', 'F', 'G', 'H'],
        'Marks': [68, 72, 80, 77]}

df = pd.DataFrame(data)

# Bar plot using seaborn
sns.barplot(x='Student', y='Marks', data=df, palette='Greens')
plt.title("Seaborn Bar Plot (Modified)")
plt.show()
```

```python
# Sample data
days = ['Sat', 'Sun', 'Mon', 'Tue', 'Wed']
temp = [30, 32, 31, 29, 28]

sns.lineplot(x=days, y=temp, marker='D', color='brown')
plt.title("Temperature Over New Days")
plt.xlabel("Days")
plt.ylabel("Temperature (C)")
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
ages = [25, 26, 27, 25, 28, 29, 27, 26, 28, 29, 30, 28]

# Create histogram
sns.histplot(ages, bins=4, color='plum', kde=True)  # kde=True adds smooth curve

# Add title and labels
plt.title("Age Distribution (Seaborn Histogram, Modified)")
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.show()
```

```python
# Sample data
x = [3, 6, 9, 6, 1, 12, 1, 8]
y = [55, 66, 77, 88, 99, 66, 77, 88]

# Create scatter plot
sns.scatterplot(x=x, y=y, color='darkgreen', s=120)  # s=120 sets size of points

# Add title and labels
plt.title("Scatter Plot Example (Seaborn, Modified)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (correlation-like table)
data = {
    'English': [65, 70, 60, 75],
    'History': [55, 68, 72, 80],
    'Geography': [70, 65, 78, 82],
    'Economics': [60, 75, 80, 77]
}

# Convert to DataFrame
df = pd.DataFrame(data, index=['Student E', 'Student F', 'Student G', 'Student H'])

# Create heatmap
sns.heatmap(df, annot=True, cmap='OrRd', linewidths=1)

# Add title
plt.title("Students Marks Heatmap (Modified)")
plt.show()
```
